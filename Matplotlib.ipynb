```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(y,x)
```

```python
plt.plot(x,y,"r")
```

```python
plt.plot(x,y,"p")
```

```python
plt.plot(x,y,"y")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Graph of X vs Y")
```

```python
plt.plot(x,y,"y")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Graph of X vs Y")
plt.show() #it will remove "Text(0.5, 1.0, 'Graph of X vs Y')"
```

```python
## Creating multiplots on same canvas 
```

```python
plt.subplot(1,2,1) #(no. rows, no. col, plot)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g--')
```

```python
plt.figure()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9]) #(left,bottom,width,height)

axes.plot(x,y,'g')
axes.set_xlabel('X values')
axes.set_ylabel('Y values')
axes.set_title('Graph of X vs Y')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.9]) #(left,bottom,width,height)
axes2 = fig.add_axes([0.2,0.5,0.3,0.4]) #(left,bottom,width,height)

axes1.plot(x,y,'g')
axes1.set_xlabel('X values')
axes1.set_ylabel('Y values')
axes1.set_title('Graph of X vs Y')

axes2.plot(y,x,'r')
axes2.set_xlabel('X values')
axes2.set_ylabel('Y values')
axes2.set_title('Graph of Y vs X')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.9]) #(left,bottom,width,height)
axes2 = fig.add_axes([1,0.5,0.3,0.4]) #(left,bottom,width,height)

axes1.plot(x,y,'g')
axes1.set_xlabel('X values')
axes1.set_ylabel('Y values')
axes1.set_title('Graph of X vs Y')

axes2.plot(y,x,'r')
axes2.set_xlabel('X values')
axes2.set_ylabel('Y values')
axes2.set_title('Graph of Y vs X')
plt.show()
```

## Subplots

```python
fig,axes = plt.subplots()

axes.plot(x,y,'g')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Graph')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig
```

```python
axes
```

```python
for v in axes:
    v.plot(x,y,'g')
    v.set_xlabel('X axis')
    v.set_ylabel('Y axis')
    v.set_title('X vs Y')
```

```python
fig
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for v in axes:
    v.plot(x,y,'g')
    v.set_xlabel('X axis')
    v.set_ylabel('Y axis')
    v.set_title('X vs Y')

fig
plt.tight_layout() #second graph will not override first graph
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)

for v in axes:
    v.plot(x,y,'g')
    v.set_xlabel('X axis')
    v.set_ylabel('Y axis')
    v.set_title('X vs Y')

fig
plt.tight_layout() #second graph will not override first graph
```

```python
fig = plt.figure(figsize=(8,4),dpi=100)
```

```python
fig,axes = plt.subplots(figsize=(8,4),dpi=100)

axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Graph')
```
